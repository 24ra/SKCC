```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## 1번 문제

sm = pd.read_excel("smoke.xlsx")
smokers = pd.DataFrame(sm)

smokers
```

```python
## 2번 문제

smokers.shape
```

```python
## 3번 문제

smokers[:10]
```

```python
## 4번 문제

female = smokers[smokers['Gender'] == 'F']
female.shape[0]
```

```python
male = smokers[smokers['Gender'] == 'M']
male.shape[0]
```

```python
## 5번 문제

smoker_yes = smokers[smokers['Smoker'] == 'yes']
smoker_yes.shape[0]
```

```python
(smoker_yes.shape[0]/smokers.shape[0])*100
```

```python
## 6번 문제

# 여성 흡연자 수

female_yes = female[female['Smoker'] == 'yes']
female_yes.shape[0]
```

```python
# 여성 흡연자 비율

(female_yes.shape[0]/female.shape[0])*100
```

```python
# 남성 흡연자 수

male_yes = male[male['Smoker'] == 'yes']
male_yes.shape[0]
```

```python
# 남성 흡연자 비율

(male_yes.shape[0]/male.shape[0])*100
```

```python
## 7번 문제

smokers['BMI'] = smokers['Weight']/(smokers['Height']*smokers['Height'])

smokers
```

```python
## 8번 문제

# 평균

smokers.groupby(['Gender']).mean().BMI
```

```python
# 표준편차

smokers.groupby(['Gender']).std().BMI
```

```python
## 9번 문제

%matplotlib inline

plot_data = smokers[['Gender', 'BMI']]

boxplot = plot_data.boxplot(by = 'Gender')
```
